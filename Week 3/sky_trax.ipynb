```python
##  importing requirements
import requests
from bs4 import BeautifulSoup

## Getting and parsing the html
url = "https://www.airlinequality.com/airline-reviews/qatar-airways"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
soup

articles = soup.find_all("article", itemprop="review")
articles
```

```python

## Getting short reviews
short_review = []
for article in articles:
    short_review.append(article.find("h2"))
short_review

for review in short_review:
    print(review.text)

```

```python
## Getting Dates
dates = []
for article in articles:
    dates.append(article.find("time"))
dates
for date in dates:
    print(date.text)

```

```python
## Getting long reviews
long_review = []
for article in articles:
    long_review.append(article.find("div", class_="text_content"))
long_review
for longs in long_review:
    print(longs.text)
```

```python
lis = zip(long_review, dates)

for lon, date in lis:
    print(f"{date.text}: {lon.text}")
```

```python
import requests
from bs4 import BeautifulSoup

page_num = 1
dates  = []
short_reviews = []
long_reviews = []

base_url= "https://www.airlinequality.com/airline-reviews/austrian-airlines/"

while True:
    url = f"{base_url}page/{page_num}"
    response2 = requests.get(url)
    soup = BeautifulSoup(response2.text, "html.parser")

    articles = soup.find_all("article", itemprop="review")

    for article in articles:
        a = article.find("time")
        dates.append(a.text)

    for article in articles:
        b = article.find("h2")
        short_reviews.append(b.text)

    for article in articles:
        c = article.find("div", class_="text_content")
        long_reviews.append(c.text)

    page_num += 1

```
